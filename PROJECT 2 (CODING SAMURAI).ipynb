LINEAR REGRESSION ON SIMPLE DATASET

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'Advertising Spend (₹1000)': [1, 2, 3, 4, 5],
    'Sales (₹1000)': [1.5, 1.9, 3.2, 3.8, 4.5]
})
X = data[['Advertising Spend (₹1000)']]
y = data['Sales (₹1000)']
model = LinearRegression()
model.fit(X, y)
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
predicted_sales = model.predict([[6]])
print(f"Predicted Sales for ₹6000 spend: {predicted_sales[0]:.2f} (₹1000s)")
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Advertising Spend (₹1000)')
plt.ylabel('Sales (₹1000)')
plt.title('Advertising Spend vs Sales')
plt.legend()
plt.show()
```

INSIGHT

```python
1. For every ₹1000 spent on ads, sales increase by ~₹0.7k.

2. At ₹0 spend, base sales are around ₹0.8k.

3. Model fits the data well.
```
